```python
df = pd.read_csv('ipl.csv')
df.head()
```

```python
df.info()
```

```python
df2 = df[[column for column in df if df[column].count() / len(df) >= 0.3]]
del df2['info.overs']
print("List of dropped columns:", end=" ")
for c in df.columns:
    if c not in df2.columns:
        print(c, end=", ")
print('\n')
df = df2
```

```python
print(df['info.outcome.by.wickets'].describe())
plt.figure(figsize=(9, 8))
sns.distplot(df['info.outcome.by.wickets'], color='g', bins=100, hist_kws={'alpha': 0.4});
```

```python
list(set(df.dtypes.tolist()))
```

```python
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.head()
```

```python
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8); # ; avoid having the matplotlib verbose informations
```

```python
df_num_corr = df_num.corr()['info.outcome.by.wickets'][:-1] # -1 because the latest row is SalePrice
golden_features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with wickets:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
for i in range(0, len(df_num.columns), 5):
    sns.pairplot(data=df_num,
                x_vars=df_num.columns[i:i+5],
                y_vars=['info.outcome.by.wickets'])
```
